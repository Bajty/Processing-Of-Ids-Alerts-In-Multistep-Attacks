# Classes for aggregation

```python
# alert
class E():
    def __init__(self):
        self.msg = None
        self.sip = None
        self.dip = None
        self.sport = None
        self.dport = None
        self.sid = None
        self.time = None
        self.proto = None
        self.tcpflags = None
        self.tos = None
    def __str__(self):
        return str(self.time) + ", " + str(self.sip) + ", " + str(self.dip) + ", " + str(self.sport) + ", " + str(self.dport) + ", " + str(self.msg) + ", " + str(self.sid)
    def __repr__(self):
        return str(self.time) + ", " + str(self.sip) + ", " + str(self.dip) + ", " + str(self.sport) + ", " + str(self.dport) + ", " + str(self.msg) + ", " + str(self.sid)

    # aggregated meta-alert
class A():
    def __init__(self):
        self.sid = set()
        self.msg = set()
        self.sip = set()
        self.dip = set()
        self.sport = set()
        self.dport = set()
        self.sip_count = 0
        self.dip_count = 0
        self.sport_count = 0
        self.dport_count = 0
        self.start_time = None
        self.end_time = None
        self.mean = None
        self.median = None
        self.std = None
        self.min = None
        self.max = None
        self.proto = set()
        self.tcpflags = set()
        self.tos = set()
        self.events = []
    def __str__(self):
        return str(self.sid) + ", " + str(self.sip) + ", " + str(self.dip) + ", " + str(self.sport) + ", " + str(self.dport) + ", " + str(self.msg) + ", " + str(self.start_time) + ", " + str(self.end_time) + ", " + str(len(self.events))
    def __repr__(self):
        return str(self.sid) + ", " + str(self.sip) + ", " + str(self.dip) + ", " + str(self.sport) + ", " + str(self.dport) + ", " + str(self.msg) + ", " + str(self.start_time) + ", " + str(self.end_time) + ", " + str(len(self.events))
    def __eq__(self, other):
        if self.sid == other.sid and self.sip == other.sip and self.dip == other.dip and self.sport == other.sport and self.dport == other.dport and self.start_time == other.start_time and self.end_time == other.end_time and self.events == other.events:
            return True
        else:
            return False
```

```python
all_events = []
```

```python
for index, row in all_data_filtered.iterrows():
    e = E()
    e.msg = row['rule_name']
    e.sip = row['from_addr']
    e.dip = row['to_addr']
    e.sport = row['from_port']
    e.dport = row['to_port']
    e.sid = row['sid']
    e.time = row['datetime']
    e.proto = row['proto']
    e.tcpflags = row['tcpflags']
    e.tos = row['tos']
    all_events.append(e)
```

```python
def time_diff(bigger, smaller):
    return pd.Timedelta(bigger - smaller).seconds
```

# Aggregation

```python
delta = 1800
agg_events = []
actual = []
for ev in all_events:
    added = False
    for av in actual:
        if (ev.time >= av.end_time and time_diff(ev.time, av.end_time) <= delta) or (ev.time >= av.start_time and ev.time <= av.end_time):
            if ((ev.sip in av.sip and len(av.sip) == 1) or (ev.dip in av.dip and len(av.dip) == 1)) and ev.sid in av.sid:
                added = True
                av.sip.add(ev.sip)
                av.msg.add(ev.msg)
                av.dip.add(ev.dip)
                av.sid.add(ev.sid)
                av.sport.add(ev.sport)
                av.dport.add(ev.dport)
                av.proto.add(ev.proto)
                av.tcpflags.add(ev.tcpflags)
                av.tos.add(ev.tos)
                av.end_time = ev.time
                av.events.append(ev)
                break
        else:
            agg_events.append(av)
            actual.remove(av)
    if not added:
        a = A()
        a.sip.add(ev.sip)
        a.msg.add(ev.msg)
        a.dip.add(ev.dip)
        a.sid.add(ev.sid)
        a.sport.add(ev.sport)
        a.dport.add(ev.dport)
        a.proto.add(ev.proto)
        a.tcpflags.add(ev.tcpflags)
        a.tos.add(ev.tos)
        a.end_time = ev.time
        a.start_time = ev.time
        a.events.append(ev)
        actual.append(a)
for av in actual:
    agg_events.append(av)
```

```python
correlated_sequences = []
aggregated_alerts = agg_events_sorted.copy()

while len(aggregated_alerts) > 0:
    actual_sequence = [aggregated_alerts.pop(0)]
    future_aggregated_alerts = []
    for i in range(len(aggregated_alerts)):
        a = aggregated_alerts.pop(0)
        added_to_actual = False
        for f in actual_sequence:
            if f.start_time <= a.start_time:
                if len(f.dip.intersection(a.sip)) >= 1 or (len(f.sip.intersection(a.sip)) >= 1 and len(f.dport.intersection(a.dport)) >=1) or (len(f.dip.intersection(a.dip)) >= 1 and len(f.dport.intersection(a.dport)) >=1):
                    added_to_actual = True
                    actual_sequence.append(a)
                    break
        if not added_to_actual:
            future_aggregated_alerts.append(a)
    aggregated_alerts = future_aggregated_alerts.copy()
    correlated_sequences.append(actual_sequence)
for c in correlated_sequences:
    print(len(c))
correlated_sequences[2]
```
